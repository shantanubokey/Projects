```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cat=df.columns[df.dtypes=='object']
con=df.columns[df.dtypes!='object']
```

### UNIVARIATE ANALYSIS

```python
plt.figure(figsize=(26,10))
df['PROD_NAME'].value_counts().plot(kind='bar',title='Count Plot')
plt.show()
```

```python
plt.figure(figsize=(18,8))
plt.subplot(2,2,1)
for i in con :
    sns.histplot(data=df,x=i,kde=True)
    plt.title(f'Histogram for {i}')
    plt.show()
```

### BIVARIATE ANALYSIS

```python
for i in con:
    if i!='TOT_SALES':
        plt.scatter(df['TOT_SALES'],df[i])
        plt.title(f' TOT_SALES VS {i}')
        plt.xlabel('TOT_SALES')
        plt.ylabel(f'{i}')
        plt.show()
```
