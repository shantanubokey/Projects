```python
def url_txt(url):
    para=[]
    data=requests.get(url).text
    soup= BeautifulSoup(data,'lxml')
    
    # Extract the title
    title_tag = soup.find('meta', {'property': 'og:title'})
    title = title_tag['content'].strip() if title_tag else ''

    # Extract paragraphs
    p=soup.find_all('p')
    for parag in p:
        parag= parag.text.strip()
        para.append(parag)
    
    # Join paragraphs into a single string
    text=' '.join(para)
    
    # Concatenate title and paragraphs
    text = title + " " + text
    
    return text
```

```python
url_txt(url)
```

### Sentimental analysis

```python
import numpy as np
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
nsw=set(stopwords.words('english'))
```

```python
def load_custom_stopwords(file_path):
    custom_stop_words = set()
    with open(file_path, 'r') as file:
        for line in file:
            custom_stop_words.add(line.strip().lower())
    return custom_stop_words
```

```python
csw=load_custom_stopwords('merge_from_ofoct.txt')
```

```python
stop_words=nsw.union(csw)
```

```python
stop_words
```

```python
from nltk.tokenize import sent_tokenize,RegexpTokenizer
```

```python
def tokenz(text):
    tokenizer=RegexpTokenizer(r'\w+')
    text=str(text)
    tokens=tokenizer.tokenize(text)
    filter=[token for token in tokens if token.lower() not in stop_words]
    
    return filter       
```

### Creating postive and negative words in use

```python
positive_words=load_custom_stopwords('positive-words.txt')
negative_words=load_custom_stopwords('negative-words.txt')
```

```python
positive_words
```

### 1. Positive score

```python
def pos_sco(text):
    tokens=tokenz(text)
    pos_sco=0
    for token in tokens:
        if token in positive_words:
            pos_sco+=1
    return pos_sco            
```

```python
df['POSITIVE SCORE']=df['text'].apply(pos_sco)
```

```python
df.head()
```

### 2. Negative score

```python
def neg_score(text):
    tokens=tokenz(text)
    neg_sco=0
    for token in tokens:
        if token in negative_words:
            neg_sco-=1
    return neg_sco *-1        
```

```python
df['NEGATIVE SCORE']=df['text'].apply(neg_score)
```

```python
df['NEGATIVE SCORE']
```

### 3. Polarity score (range from 0 to 1)

```python
def pol_sco(pos_sco,neg_score):
    pol_score=(pos_sco-neg_score)/(pos_sco+neg_score+0.000001)
    return pol_score
```

```python
df['POSITIVE SCORE'][0:5]
```

```python
df['Polarity score'] = df.apply(lambda x: pol_sco(x['POSITIVE SCORE'], x['NEGATIVE SCORE']), axis=1)
```

```python
df['Polarity score'][0:5]
```

### 4. Subjectively score

```python
def sub_score(pos_sco,neg_score,text):
    tokens=tokenz(text)
    sub_s=(pos_sco+neg_score)/((len(tokens))+0.000001)
    return sub_s
```

```python
df['SUBJECTIVE SCORE']=df.apply(lambda x: sub_score(x['POSITIVE SCORE'],x['NEGATIVE SCORE'],x['text']),axis=1)
```

```python
df['SUBJECTIVE SCORE'][0:5]
```

### 5. Average sentence length

```python
def avg_sen_len(text):
    words=len(tokenz(text))
    sentence=len(text.split('.'))
    avg_len=words/sentence
    return avg_len
```

```python
df['AVERAGE SENTENCE LENGTH']=df['text'].apply(avg_sen_len)
```

```python
df['AVERAGE SENTENCE LENGTH'][0:5]
```

### Percentage of complex words

```python
def per_complex(text):
    tokens=tokenz(text)
    complex_cnt=0
    for token in tokens:
        vowels=0
        if token.endswith(('es','ed')):
            pass
        else:
            for l in token:
                if (l=='a'or l=='e'or l=='i' or l=='o' or l=='u'):
                    vowels+=1
            if vowels>2:
                complex_cnt+=1
    if len(token)!=0:
        return complex_cnt/len(tokens)
```

```python
df['PERCENTAGE OF COMPLEX WORDS']=df['text'].apply(per_complex)
```

### FOG INDEX

```python
def fog_index(avg_sen_len,percentage_complex_words):
    return 0.4*(avg_sen_len+percentage_complex_words)
```

```python
df['FOG index']=df.apply(lambda x: fog_index(x['AVERAGE SENTENCE LENGTH'],x['PERCENTAGE OF COMPLEX WORDS']),axis=1)
```

```python
def compplex_word_count(text):
    tokens=tokenz(text)
    complex_count=0
    for token in tokens:
        vowels=0
        if token.endswith(('es','ed')):
            pass
        else:
            for l in token:
                if(l=='a' or l=='e' or l=='i' or l=='o' or l=='u'):
                    vowels+=1
            if vowels>2:
                complex_count+=1
    return complex_count      
```

### Syllable per word

```python
def syllable_per_word(text):
    words=tokenz(text)
    words_cnt=len(words)
    vowels=0
    for word in words:
        if word.endswith(('es','ed')):
            pass
        else:
            for l in word:
                if(l=='a' or l=='e' or l=='i' or l=='o' or l=='u'):
                    vowels+=1
    return vowels/words_cnt
```

```python
df['SYLLABLE PER WORD']=df['text'].apply(syllable_per_word)
```

### Personal Pronouns

```python
import re
```

```python
def calculate_personal_pronoun(text):
    words=tokenz(text)
    pronoun_re=r'\b(I|my|we|us|ours)\b'
    matches=re.findall(pronoun_re,text)
    return len(matches)
```

```python
df['PERSONAL PRONOUN']=df['text'].apply(calculate_personal_pronoun)
```

### Average word length

```python
def avg_word_length(text):
    words=tokenz(text)
    charcnt=0
    for word in words:
        charcnt+=len(word.strip())
    if len(words)!=0:
        return charcnt/len(words)    
```

```python
df['AVERAGE WORD LENGTH']=df['text'].apply(avg_word_length)
```
